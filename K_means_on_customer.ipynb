```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("customer_segmentation.csv")
```

```python
df
```

```python
from sklearn.cluster import KMeans
X=df[['Income_k',"Spending_Score"]]
model=KMeans(n_clusters=2,random_state=42,n_init=10)
model.fit(X)


df['Cluster'] = model.labels_
```

```python
df['Group']=model.fit_predict(X)
```

```python

plt.figure(figsize=(6,5))
for i in df['Group'].unique():
    group_data=df[df['Group']==i]
    plt.scatter(group_data['Income_k'],group_data['Spending_Score'])

plt.xlabel('Income_k')
plt.ylabel('Spending Score')
plt.title('Customer Segments (K-Means)')
plt.legend()
plt.show()
```

```python
new_customer = [[20, 90]]  
cluster_label = model.predict(new_customer)
print("New customer belongs to Cluster:", cluster_label[0])
```
